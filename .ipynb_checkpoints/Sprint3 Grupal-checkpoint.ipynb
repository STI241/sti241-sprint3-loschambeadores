```python
import pandas as pd
from sklearn.neighbors import NearestNeighbors
from sklearn.metrics.pairwise import cosine_similarity
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA

df = pd.read_csv("Sprint3.csv", index_col=0)
```

```python
def calcular_distancia_cosine_entre_todos(df):
    distancias = {}
    for usuario1 in df.columns:
        distancias_usuario1 = {}
        vector_usuario1 = df.loc[:, usuario1].values.reshape(1, -1)
        for usuario2 in df.columns:
            if usuario1 != usuario2:
                vector_usuario2 = df.loc[:, usuario2].values.reshape(1, -1)
                distancia_cosine = cosine_similarity(vector_usuario1, vector_usuario2)
                distancias_usuario1[usuario2] = distancia_cosine[0][0]
        distancias[usuario1] = distancias_usuario1
    return distancias

distancias_entre_usuarios = calcular_distancia_cosine_entre_todos(df)

for usuario, distancias_usuario in distancias_entre_usuarios.items():
    print("Distancias cosine para el usuario", usuario, ":", distancias_usuario)
```

```python
from sklearn.neighbors import NearestNeighbors

def vecinos_cercanos(usuario, K, df):
    matriz_caracteristicas = df.T.values  
    
    modelo = NearestNeighbors(n_neighbors=K+1, algorithm='auto')  
    
    modelo.fit(matriz_caracteristicas)
    
    distancia, indices = modelo.kneighbors([df.loc[:, usuario].values])  
    
    vecinos = df.columns[indices[0][1:]].tolist()  
    
    distancias = distancia[0][1:]
    
    return vecinos, distancias

def algoritmo_recomendacion(K, df):
    recomendaciones = {}
    for usuario in df.columns:
        vecinos, distancias = vecinos_cercanos(usuario, K, df)
        recomendaciones[usuario] = {"vecinos": vecinos, "distancias": distancias}
    return recomendaciones

K = 3  

recomendaciones = algoritmo_recomendacion(K, df)

for usuario, info_recomendacion in recomendaciones.items():
    print("Vecinos más cercanos a", usuario, ":", info_recomendacion["vecinos"])
    print("Distancias correspondientes:", info_recomendacion["distancias"])
    print()
```

```python
def visualizar_datos(df):
    pca = PCA(n_components=2)
    datos_pca = pca.fit_transform(df.T) 
    
    plt.figure(figsize=(10, 6))
    plt.scatter(datos_pca[:, 0], datos_pca[:, 1], alpha=0.5)
    plt.title('Visualización de Datos en Espacio Reducido (PCA)')
    plt.xlabel('Ingredientes')
    plt.ylabel('Preferencia')
    for i, usuario in enumerate(df.columns):  
        plt.annotate(usuario, (datos_pca[i, 0], datos_pca[i, 1]))
    plt.show()

visualizar_datos(df)
```
